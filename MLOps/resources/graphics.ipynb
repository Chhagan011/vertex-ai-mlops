## Helper Funtions: For Mermaid Plots

Designed to render plots from encoded strings sent to `https://mermaid.ink` after [this tutorial](https://mermaid.js.org/ecosystem/tutorials.html#jupyter-integration-with-mermaid-js)

```python
def mermaid_render(graph, mainpath = mainpath, filepath = None):
    encode = base64.b64encode(graph.encode('utf-8')).decode('ascii')
    response = requests.get('https://mermaid.ink/img/' + encode)
    if filepath:
        # verify director exists:
        fullpath = mainpath+filepath
        directory = os.path.dirname(fullpath)
        if directory and not os.path.exists(directory):
            os.makedirs(directory)

        # save file
        with open(fullpath, 'wb') as f:
            f.write(response.content)
        
    return Image(response.content)
```
