## Helper Funtions: For Mermaid Plots

Designed to render plots from encoded strings sent to `https://mermaid.ink` after [this tutorial](https://mermaid.js.org/ecosystem/tutorials.html#jupyter-integration-with-mermaid-js)

```python
def mermaid_render(graph, mainpath = mainpath, filepath = None, remote = True):
    if filepath:
        # verify director exists:
        fullpath = mainpath+filepath
        directory = os.path.dirname(fullpath)
        if directory and not os.path.exists(directory):
            os.makedirs(directory)
    
    if remote: # use mermaid.ink/img to do the render
        encode = base64.b64encode(graph.encode('utf-8')).decode('ascii')
        response = requests.get('https://mermaid.ink/img/' + encode)
        if filepath:
            # save file
            with open(fullpath, 'wb') as f:
                f.write(response.content)
        return Image(response.content)
    else: # use local mermaid-cli (mmdc)
    #    if filepath:
    #        with open(fullpath+'.mdd', 'w') as f:
    #            f.write(graph)
    #        subprocess.run(["mmdc", "-i", fullpath+'.mdd', "-o", fullpath]) # -o is output to file, -t for theme, -b for background, -i is the input graph
    #        os.remove(fullpath+'.mdd')
    #    else: print('The `filepath` parameter is required for local processing (remote = False).')
        print('Still working on headless chrome in notebook')
        return #Image(fullpath)
```
